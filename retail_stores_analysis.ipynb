```python
#to handle null values
df = pd.read_csv('orders.csv',na_values = ['Not Available','unknown'])
```

```python
#rename the column names 
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
#create a new columns of discounts for feature engineering
df['discounts'] = df['list_price']*df['discount_percent']*.01
```

```python
 df['sale_price'] = df['list_price']-df['discounts']
```

```python
df['profit']=df['sale_price']-df['cost_price']
```

```python
#convert the data type of the order_date column 
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
#drop the unnecessary columns from the dataset
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```
